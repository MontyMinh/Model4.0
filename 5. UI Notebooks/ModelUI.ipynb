```python
class ProgramError(Exception):
    """Class for Runtime Error"""
    
    def __init__(self):
        super().__init__("An error has occured while optimizing the program\n              See the error message above and below for more details")

class UI:
    def exit_message():
        print('Process Terminated')

    @classmethod
    def open_file(cls):
        """Function to collect the file path"""

        filepath = askopenfile(mode='r', filetypes=[('Excel Files', '*xlsx, *xlsm')])

        # If doesn't exist, display retry message
        if filepath is None:
            # Destroy message box if exist
            try:
                cls.msg.destroy()
            except:
                pass
            cls.msg = Label(cls.root,
                            text='No File Uploaded, Retry!',
                            foreground='red')
            cls.msg.grid(row=4, columnspan=3, pady=10)

        # If it does, display success message
        else:
            # Destroy message box if exist
            try:
                cls.msg.destroy()
            except:
                pass

            # Format filepath
            last_str_index = filepath.name[::-1].index('/')
            cls.filepath = filepath.name
            cls.msg = Label(
                cls.root,
                text=
                f'{cls.filepath[-last_str_index:]} \nUploaded Successfully!',
                foreground='green')
            cls.msg.grid(row=4, columnspan=3, pady=10)
            cls.upload_button['text'] = '  Reupload  '

            # Store cls.filepath in Data
            Data.filepath = cls.filepath

    @classmethod
    def run_program(cls):
        """Run optimization program"""
        
        program_text = Label(cls.root, text = "Running program...")
        program_text.grid(row = 6, columnspan = 3)
        
        # Try Execute
        try:
            execute()
        # If failed, display error message and raise Exception
        except:
            print('Failed')
            program_text = Label(
                cls.root,
                text=
                "An error has occured. See the\nterminal for more information",
                foreground='red')
            program_text.grid(row=6, columnspan=3)
                       
            # Raise Exceptions
            raise ProgramError()
        else:
            program_text = Label(
                cls.root,
                text="Program optimized and saved successfully!",
                foreground='green')
            program_text.grid(row=6, columnspan=3)

    @classmethod
    def start(cls):
        """Start the User Interface to Upload File"""
        cls.root = Tk()
        cls.root.title('Optimization Model')
        cls.root.geometry('500x250')

        # Upload input file
        upload_text = Label(cls.root, text='Upload Data File')
        upload_text.grid(row=0, column=0, padx=10)

        cls.upload_button = Button(
            cls.root,
            text='Upload File ',
            command=lambda: cls.open_file(),
        )
        cls.upload_button.grid(row=0, column=1)

        upload_message = Label(cls.root, text='',
                               foreground='red').grid(row=4,
                                                      columnspan=2,
                                                      pady=10,
                                                      padx=10)

        # Run optimization and save results
        Results.save_location = 'C:/Users/monty.minh/Downloads/Results.xlsx'

        # Program prompt
        run_text = Label(cls.root, text="Optimize Program")
        run_text.grid(row=5, column=0, padx=10)

        # Execute button
        run_button = Button(cls.root, text='Execute', command=cls.run_program)
        run_button.grid(row=5, column=1)

        cls.root.mainloop()

        cls.exit_message()

UI.start()
```
